```python
# import the packages that will be used in this project

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

# matplotlib graphs will be included next to the code
%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None



# import the csv file data using pandas
df = pd.read_csv(r'C:\Users\rielp\Downloads\movies.csv')
```

```python
#look if the data has been uploaded, as well as the format
df.head(3)
```

```python
# check for any missing data
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# look at column data types
df.dtypes
```

```python
# check for outliers
df.boxplot(column=['gross'])
```

```python
# remove any duplicates
df.drop_duplicates()
```

```python
# order the data a bbit better
df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# plot data and with a linear regression model fit using seaborn
# compare gross vs budget
sns.regplot(x="gross", y="budget", data=df)
```

```python
# compare score vs gross
sns.regplot(x="score", y="gross", data=df)
```

```python
# Correlation Matrix between all numeric columns
df.corr(method ='pearson')
```

```python
df.corr(method ='kendall')
```

```python
df.corr(method ='spearman')
```

```python
# illustrate the correlation matrix by means of a heatmap

correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix between each column")
plt.xlabel("Movie features")
plt.ylabel("Movie features")

plt.show()
```

```python
# Using factorize - assigns a random numeric value for each unique categorical value

df.apply(lambda x: x.factorize()[0]).corr(method='pearson')
```

```python
# use heat map to illustrate the correlation between each variable

correlation_matrix = df.apply(lambda x: x.factorize()[0]).corr(method='pearson')
sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix between each column")
plt.xlabel("Movie features")
plt.ylabel("Movie features")

plt.show()
```

```python
# stack each feature and the value it is correlated with

correlation_mat = df.apply(lambda x: x.factorize()[0]).corr()
corr_pairs = correlation_mat.unstack()

print(corr_pairs)
```

```python
sorted_pairs = corr_pairs.sort_values(kind="quicksort")
print(sorted_pairs)
```

```python
# show high correlation (> 0.5)

strong_pairs = sorted_pairs[abs(sorted_pairs) > 0.5]
print(strong_pairs)
```

```python
# numerize the object columns [potential better way to determine corr between an object and numerical value]
df_numerized = df

#change object type columns to categoy type
# use cat codes to assign it a numerical value
for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name]= df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized
```

```python
#redo the different methods of correlations

df_numerized.corr(method='pearson')
```

```python
# re-issue heat map to illustrate new correlations

correlation_matrix = df_numerized.corr(method='pearson')
sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Movies")
plt.xlabel("Movie features")
plt.ylabel("Movie features")

plt.show()
```

```python
#check the correlated data
df_numerized.corr()
```

```python
#unstack the correlation per variable [each variable has a list with the corr between each column]
correlation_mat = df_numerized.corr()
corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
# pair the corr values [each column is grouped]

sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
#list the corr that is above 0.25

high_corr = sorted_pairs[sorted_pairs > 0.25]
high_corr
```
